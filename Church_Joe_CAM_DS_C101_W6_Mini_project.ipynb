## Import Required Libraries and Load Data

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage
from scipy.stats import zscore


data = pd.read_csv('CUSTOMERS_CLEAN.csv')

# display first few rows of data
print(data.head())


# display shape of data
print(data.shape)
```

## Data pre-processing

### Handling missing values

```python
# Checking for missing values
print("Missing Values:\n", data.isnull().sum())

# Checking for duplicate rows
print("Duplicate Rows:", data.duplicated().sum())

print("\n" + "="*40 + "\n")

# Dropping duplicate rows - dupliucates must match across all columns
data = data.drop_duplicates()

# Dropping columns that are not crucial for task or are PII/sensitive + have missing values
data = data.drop(columns=['City', 'Postal_Code', 'State_Province'])

# Confirm no missing values remain
missing_values_after = data.isnull().sum()
print("Missing Values After Dropping Columns:\n", missing_values_after)

print("\n" + "="*40 + "\n")

print(data['Customer ID'].nunique())
```

### Cleaning important features

```python
# Function to clean currency columns by removing symbols and converting to numeric
def clean_currency(x):
    if isinstance(x, str):
        return float(x.replace('$', '').replace(',', '').replace('(', '-').replace(')', '').replace(' ', ''))
    return float(x)

# Apply the clean_currency function to relevant columns
data['Total Revenue'] = data['Total Revenue'].apply(clean_currency)
data['Unit Cost'] = data['Unit Cost'].apply(clean_currency)
data['Profit'] = data['Profit'].apply(clean_currency)

# Converting the discount to a float using lambda function
data['Discount'] = data['Discount'].apply(
    lambda x: 0.0 if x == '.' else float(x.replace('%', '').strip()) / 100 if x.replace('%', '').replace('.', '').strip().isdigit() else 0.0
)

# Confirm that the columns are now numeric
print(data[['Total Revenue', 'Unit Cost', 'Profit', 'Discount']].dtypes)

print("\n" + "="*40 + "\n")

# Display the first few rows to confirm cleaning
print(data[['Total Revenue', 'Unit Cost', 'Profit', 'Discount']].head())

print("\n" + "="*40 + "\n")

# calulcate unique customer IDs to check post aggregation:
unique_customer_ids = data['Customer ID'].nunique()
print("Number of unique customer IDs:", unique_customer_ids)

print("\n" + "="*40 + "\n")

print(data.shape)

# Convert date columns to datetime format
data['Order_Date'] = pd.to_datetime(data['Order_Date'], format='%d%b%Y', errors='coerce')
data['Delivery_Date'] = pd.to_datetime(data['Delivery_Date'], format='%d%b%Y', errors='coerce')
data['Customer_BirthDate'] = pd.to_datetime(data['Customer_BirthDate'], format='%d%b%Y', errors='coerce')
```

### Handling Outliers

After doing both before and after, it was decided that very extreme outliers should be handled in the original dataset prior to aggregation as this reduced the affect on the data going into the model.

Given the nature of e-commerce and sales data, we naturally see extremely left skew in purchase data that represents very real but rare customer behaviour. I wanted to retain these patterns in the data without affecting the visualisations too much.


```python
# print shape of data prior to outlier handling:
print(data.shape)
print("\n" + "="*40 + "\n")

# Remove any rows with negative values in the specified columns(impossible)
data = data[(data['Quantity'] >= 0) &
            (data['Total Revenue'] >= 0) &
            (data['Unit Cost'] >= 0)
]

# Function for removing very extreme outliers based on percentiles
def remove_outliers(df, columns, lower_percentile=0.001, upper_percentile=0.999):
    for col in columns:
        lower = df[col].quantile(lower_percentile)
        upper = df[col].quantile(upper_percentile)
        df = df[(df[col] >= lower) & (df[col] <= upper)]
    return df

# Apply the function to Data
numerical_columns = ['Quantity', 'Total Revenue', 'Unit Cost', 'Profit']
original_count = data.shape[0]

# Remove outliers
data = remove_outliers(data, numerical_columns)

# Show statistics
filtered_count = data.shape[0]
print(f"Original data points: {original_count}")
print(f"Data points after removing outliers: {filtered_count}")
print(f"Percentage of data retained: {filtered_count / original_count * 100:.2f}%")

# Print statistics
print(f"Data points after light outlier handling: {data.shape}")
```

So we have retained 99.39% of the data. Moving the threshold down for extreme outliers to 0.01 and 0.99, only 93% of the observations were kept which would have been far too much for this purpose.

## Feature Engineering

### Feature creation and date handling

**Note**: the recency and customer age features are both calculated assuming we are at the end of the period assessed in the data i.e. the model would be producing customer segments based on the last 5 years of data assuming we are at the last order date in 2016 in the datatset.

```python
# Calculating Recency (days since the last purchase)
max_date = data['Order_Date'].max()
recency = data.groupby('Customer ID')['Order_Date'].max().reset_index()
recency.columns = ['Customer ID', 'LastOrderDate']
recency['Recency'] = (max_date - recency['LastOrderDate']).dt.days

# Dropping LastOrderDate column as it is not needed
recency = recency.drop(columns=['LastOrderDate'])

# Calculating Customer_Age at the time of the last order in the dataset
data['Customer_Age'] = (max_date - data['Customer_BirthDate']).dt.days // 365
```

### Aggregating the dataframe to the customer level

```python
# Aggregating data to one customer per row
df_agg = data.groupby('Customer ID').agg({
    'Order ID': 'count',  # Frequency
    'Total Revenue': 'sum',  # CLV
    'Unit Cost': 'mean',  # Avg_Unit_Cost
    'Profit': 'sum',  # Total Profit
    'Customer_BirthDate': 'first'  # Birthdate (to calculate age at end of dataset)
}).reset_index()

# Rename columns
df_agg.columns = ['Customer ID', 'Frequency', 'CLV', 'Avg_Unit_Cost', 'Total_Profit', 'Birthdate']

# Merging with Recency and Age features
df_agg = df_agg.merge(recency, on='Customer ID')
df_agg = df_agg.merge(data[['Customer ID', 'Customer_Age']].drop_duplicates(), on='Customer ID')
```

### Final cleaning and export for inspection

```python
# Columns to keep
columns_to_keep = ['Customer ID', 'Frequency', 'Recency', 'CLV', 'Avg_Unit_Cost', 'Customer_Age']

# Filter the DataFrame
df_agg = df_agg[columns_to_keep]

# Checking for NaN values
print(df_agg.isnull().sum())

print("\n" + "="*40 + "\n")

# Dropping rows with NaN values
df_agg = df_agg.dropna()

# Confirming no more NaN values
print(df_agg.isnull().sum())

# Export the DataFrame to a CSV file
df_agg.to_csv('customer_data.csv', index=False)

# first few rows of final aggregated dataframe
print(df_agg.head())
print(df_agg.shape)

# Copy of aggregated dataframe and rename
customer_data = df_agg.copy()
```

## Determining Optimal Number of Clusters (k) Using Elbow and Silhouette Method

### Initialising the columnar transformer pipeline

```python
# Dropping the Customer ID column
features = customer_data.copy()
features = features.drop(columns=['Customer ID'])

# Defining the columns explcitly
feature_columns = features.columns.tolist()  # List of column names

# Defining the ColumnTransformer to scale numerical features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), feature_columns)
    ])

# Defining the pipeline with ColumnTransformer and KMeans
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('kmeans', KMeans(n_init=10))
])
```

### Performing the elbow method

```python
# Elbow method to find the optimal number of clusters
wcss = []

for k in range(2, 15):
    pipeline.set_params(kmeans__n_clusters=k)
    pipeline.fit(features)
    wcss.append(pipeline.named_steps['kmeans'].inertia_)

# Plot the Elbow Method
plt.figure(figsize=(10, 5))
plt.plot(range(2, 15), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of clusters (k)')
plt.ylabel('WCSS')
plt.show()
```

### Performing the Silhouette score method

```python
# Silhouette analysis for different numbers of clusters
range_n_clusters = range(3, 7)
silhouette_scores = []

for n_clusters in range_n_clusters:
    pipeline.set_params(kmeans__n_clusters=n_clusters)
    cluster_labels = pipeline.fit_predict(features)
    silhouette_avg = silhouette_score(features, cluster_labels)
    silhouette_scores.append(silhouette_avg)
    print(f"For n_clusters = {n_clusters}, the average silhouette_score is : {silhouette_avg}")

#  DataFrame to display the silhouette scores neatly
silhouette_df = pd.DataFrame({
    'Number of Clusters (k)': list(range_n_clusters),
    'Silhouette Score': silhouette_scores
})

print("\nSilhouette Scores for Different k Values:")
print(silhouette_df)

# plot the silhouette scores
plt.figure(figsize=(10, 5))
plt.plot(range_n_clusters, silhouette_scores, marker='o', color='orange')
plt.title('Silhouette Scores for Different k Values')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Silhouette Score')
plt.show()
```

**Interpretation**: The highest silhouette score is shown at 3 clusters with the next best coming it at 4 clusters. From the elbow diagram, it seems there is the sharpest 'elbow' point at 4 clusters.

## Performing Dimensionality Reduction with PCA & t-SNE and visualising the results

```python
# Fiting the pipeline to scale the features
pipeline.fit(features)

# Accessing the scaled features from the pipeline
scaled_features = pipeline.named_steps['preprocessor'].transform(features)

# Undergoing PCA
pca = PCA(n_components=2)
pca_components = pca.fit_transform(scaled_features)

# DataFrame with the PCA components and the cluster labels
pca_df = pd.DataFrame(data=pca_components, columns=['PC1', 'PC2'])
pca_df['Cluster'] = customer_cluster['Cluster_3']

# Plotting the PCA-reduced data
plt.figure(figsize=(10, 7))
sns.scatterplot(x='PC1', y='PC2', hue='Cluster', data=pca_df, palette='viridis', s=100)
plt.title('PCA of Customers (3 Clusters)')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```

### Observing the principal component make-up

```python
# Now access the components and explained variance
components = pca.components_
explained_variance = pca.explained_variance_ratio_

# DataFrame to show how much each original feature contributes to each principal component
pca_df = pd.DataFrame(components, columns=feature_columns)
pca_df.index = [f'PC{i+1}' for i in range(pca_df.shape[0])]
pca_df['Explained Variance'] = explained_variance

# Display the DataFrame
print("\nPrincipal Components and their Composition:")
print(pca_df)

# Plotting the explained variance
plt.figure(figsize=(10, 5))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.7, align='center')
plt.title('Explained Variance by Each Principal Component')
plt.xlabel('Principal Component')
plt.ylabel('Explained Variance Ratio')
plt.show()
```

**Interpretation**:

The first principal component (PC1) explains nearly 45% of the variance, while the second principal component (PC2) explains about 20% together accounting for approx 65% of the variance in the data. This indicates that these two components capture a significant portion of the information in the original dataset, making the **PCA plot a reliable representation of the clusters**.

**Cluster 2:** Moderately valuable, moderately frequent middle-aged customers who are more recent and have moderate to high spending per unit.

**Cluster 3:** Older, high unit-cost customers with lower engagement, older recency, and lower overall frequency.

### Perform Dimension Reduction with t-SNE

Again, we are sampling the feature set due to incredibly long compute times and crashing etc.

```python
# Sample a smaller portion of the data from the original features dataframe
small_df = features.sample(n=10000, random_state=42).reset_index(drop=True)
small_labels = small_df['Cluster_3']  # Extract the corresponding cluster labels

# Dropping the cluster columns before scaling
small_df_scaled = small_df.drop(columns=['Cluster_3'])

# Scale the sampled features
small_df_scaled = pipeline.named_steps['preprocessor'].transform(small_df_scaled)

# Perform t-SNE on the scaled data
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
tsne_components = tsne.fit_transform(small_df_scaled)

# Dataframe with the t-SNE components and the cluster labels
tsne_df = pd.DataFrame(data=tsne_components, columns=['t-SNE1', 't-SNE2'])
tsne_df['Cluster'] = small_labels.reset_index(drop=True)

# Ploting t-SNE-reduced data
plt.figure(figsize=(10, 7))
sns.scatterplot(x='t-SNE1', y='t-SNE2', hue='Cluster', data=tsne_df, palette='viridis', s=100)
plt.title('t-SNE of Customers (3 Clusters)')
plt.xlabel('t-SNE Component 1')
plt.ylabel('t-SNE Component 2')
plt.show()
```
